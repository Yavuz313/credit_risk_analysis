1. **ID**: A unique identifier assigned to each customer.  
2. **Customer_ID**: The unique identification number of the customers.  
3. **Month**: The month in which the data was collected or recorded.  
4. **Name**: The customer's name.  
5. **Age**: The customer's age.  
6. **SSN**: Social Security Number (unique identifier).  
7. **Occupation**: The customer's occupation.  
8. **Annual_Income**: The customer's annual income.  
9. **Monthly_Inhand_Salary**: The customer's monthly take-home salary.  
10. **Num_Bank_Accounts**: The number of bank accounts the customer has.  
11. **Num_Credit_Card**: The number of credit cards the customer holds.  
12. **Interest_Rate**: The annual interest rate on the customer's loans.  
13. **Num_of_Loan**: The total number of loans the customer has taken.  
14. **Type_of_Loan**: The types of loans the customer has taken.  
15. **Delay_from_due_date**: The number of days a payment has been delayed after the due date.  
16. **Num_of_Delayed_Payment**: The number of delayed payments.  
17. **Changed_Credit_Limit**: The rate of changes made to the customer's credit limit.  
18. **Num_Credit_Inquiries**: The number of credit inquiries made regarding the customer.  
19. **Credit_Mix**: The balance of different types of credit held by the customer.  
20. **Outstanding_Debt**: The total unpaid debt of the customer.  
21. **Credit_Utilization_Ratio**: The credit utilization ratio (based on the available credit limit).  
22. **Credit_History_Age**: The length of the customer's credit history (in years and months).  
23. **Payment_of_Min_Amount**: Indicates whether the customer made the minimum payment.  
24. **Total_EMI_per_month**: The total Equated Monthly Installment (EMI) paid each month.  
25. **Amount_invested_monthly**: The amount the customer invests each month.  
26. **Payment_Behaviour**: Information regarding the customer's payment behavior.  
27. **Monthly_Balance**: The balance the customer has at the end of each month.  
28. **Credit_Score**: The customer's credit score (dependent variable).

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
def eval_metric(model, X_train, y_train, X_test, y_test):
    y_train_pred_probabilities = model.predict(X_train)
    y_train_pred = y_train_pred_probabilities.argmax(axis=1)
    y_pred_probabilities = model.predict(X_test)
    y_pred = y_pred_probabilities.argmax(axis=1)
    
    print("Test Set:")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    
    print("\nTrain Set:")
    print(confusion_matrix(y_train, y_train_pred))
    print(classification_report(y_train, y_train_pred))
```

```python
def eval_metric(model, X_train, y_train, X_test, y_test):
    y_train_pred_probabilities = model.predict(X_train)
    y_train_pred = y_train_pred_probabilities.argmax(axis=1)
    y_pred_probabilities = model.predict(X_test)
    y_pred = y_pred_probabilities.argmax(axis=1)
    
    print("Test Set:")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    
    print("\nTrain Set:")
    print(confusion_matrix(y_train, y_train_pred))
    print(classification_report(y_train, y_train_pred))
```

```python
# List of features to be released
features_to_drop = [
    'Credit_Utilization_Ratio', 
    'Total_EMI_per_month', 
    'Num_of_Loan',
    'Annual_Income', 
    'Amount_invested_monthly',
    'Monthly_Inhand_Salary', 
    'Age', 
    'Monthly_Balance'
]

# Credit_Score is allocated as target variable
y = df["Credit_Score"]

# We extract the specified features from the dataset X
X = df.drop(["Credit_Score"] + features_to_drop, axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Early stopping
early_stop = EarlyStopping(monitor='val_loss', patience=70, restore_best_weights=True) # val_loss u dene
```

```python
from tensorflow.keras.layers import Dropout
from keras.layers import BatchNormalization

model = Sequential([

    Dense(units=512, activation="tanh"),
    BatchNormalization(),
    Dropout(0.3),
    
    Dense(units=256, activation="relu"),
    BatchNormalization(),
    Dropout(0.2),  # Dropout ekledik
    
    Dense(units=128, activation="relu"),
    BatchNormalization(),
    Dropout(0.2),

    Dense(units=128, activation="relu"),
    BatchNormalization(),
    
    Dense(units=128, activation="relu"),
    BatchNormalization(),
    Dropout(0.2),  # Dropout ekledik
    
    Dense(units=64, activation="relu"),
    BatchNormalization(),
    
    Dense(units=3, activation="softmax")
])


opt = Adam(learning_rate=0.001)

model.compile(optimizer=opt, loss="sparse_categorical_crossentropy", metrics=["accuracy"])

history = model.fit(x=X_train, y=y_train, 
                    validation_data=(X_test, y_test),
                    batch_size=256, epochs=700,
                   callbacks=[early_stop])
```
